### Importación de datos



```python
import pandas as pd
import matplotlib.pyplot as plt

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



##**Venta total por cada tienda**

1.   Elemento de lista
2.   Elemento de lista



```python
suma_precios_tienda = tienda["Precio"].sum()
suma_precios_tienda2 = tienda2["Precio"].sum()
suma_precios_tienda3 = tienda3["Precio"].sum()
suma_precios_tienda4 = tienda4["Precio"].sum()
```

```python
print(f"La suma total de la columna 'precio' en la tienda 1 es: {suma_precios_tienda}")
print(f"La suma total de la columna 'precio' en la tienda 2 es: {suma_precios_tienda2}")
print(f"La suma total de la columna 'precio' en la tienda 3 es: {suma_precios_tienda3}")
print(f"La suma total de la columna 'precio' en la tienda 4 es: {suma_precios_tienda4}")
```

##**Gráfica del porcentaje de ventas de las tiendas**

```python
#Lista de nombres de las tiendas
tiendas = ["Tienda 1", "Tienda 2", "Tienda 3", "Tienda 4"]
#Valores (totales de precios)
valores = [suma_precios_tienda, suma_precios_tienda2, suma_precios_tienda3, suma_precios_tienda4]
#Crear gráfica de pastel
plt.figure(figsize=(6,6)) #Tamaño visual de la gráfica
explode = [0.05 if v == max(valores) else 0 for v in valores] #Para destacar la Tienda con mayor porcentaje
plt.pie(valores, labels=tiendas, autopct='%1.1f%%', startangle=90, explode=explode) #Creación de la gráfica de pastel
plt.title("Porcentaje de ventas por tienda", fontsize=14) #Cambiando el título
plt.axis('equal')  #Hace que el círculo no se deforme
plt.show()
```
